```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('GWSI_WW_LEVELS.csv', index_col=0,)
df['WLWA_MEASUREMENT_DATE'] = pd.to_datetime(df['WLWA_MEASUREMENT_DATE'])
site_list = df.index.unique()
print(site_list)
no_sites = len(site_list)
print(no_sites)
df
```

```python
for site in site_list:
   sub = df.loc[site]
   sub.index = sub['WLWA_MEASUREMENT_DATE']
   sub['WLWA_DEPTH_TO_WATER'].plot()

plt.legend()
```
